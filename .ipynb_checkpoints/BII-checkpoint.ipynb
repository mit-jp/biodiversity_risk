```python
import pandas as pd
import matplotlib.pyplot as plt
import json
from itertools import chain
```

```python

df = pd.read_csv("BII.csv",na_values=['nan'])
df = df.dropna(axis=1, how='all', thresh=None, subset=None, inplace=False)
for i in range(len(df)):
    df.loc[i,"region"]=str(df.loc[i,"area_code"])[-3:]
print(df)
```

```python
with open('area_code_hier.json', 'r') as f:
  data = json.load(f)
def explore(list,dict):
    print(list)
    print(5*"\ n")
    if list['children']==[]:
        dict.update({list["label"]: list["value"]})
        return dict
    else:
        for k in range(len(list["children"])):
            dict.update(explore(list["children"][k],dict))
        return dict

print(explore(data["children"],{}))
```

```python
import pycountry

mapping = {country.name: country.alpha_3 for country in pycountry.countries}
df["code"]=df['region_name'].map(mapping)
df = df[df['scenario'] == 2021]
df = df[df['value'] >= 0]
df = df[df['long_goal'] == "Index"]
df = df[df['dimension'] == "score"]
df.dropna()
print(df)
```

```python
df=df.dropna()
```

```python
df2 = df[df['year'] == 2021]
```

```python
import plotly.io as pio
#pio.renderers.default = "browser"

import plotly.express as px
fig = px.choropleth(df2,locations="region",
                    color="value",
                    hover_name="region",
                    #animation_frame=df["scenario"],
                    color_continuous_scale="Viridis",
                    range_color=(0, 1),scope="africa"
                    )
#fig["layout"].pop("updatemenus")
fig.show()
fig.write_image("ohi_africa.pdf")
```
